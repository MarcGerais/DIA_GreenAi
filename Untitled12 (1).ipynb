```python
import glob
import pandas as pd
import matplotlib.pyplot as plt

# --- Chargement ---
print("=== Chargement des données ===")
files = glob.glob("/content/global-gridded-annual-glacier-mass-change_*.csv")
df = pd.concat((pd.read_csv(f) for f in files), ignore_index=True)

print(f"Nombre de fichiers trouvés : {len(files)}")
print(f"Nombre total de lignes : {len(df)}")
print()


# --- NaN / valeurs manquantes ---
print("=== Vérification des valeurs manquantes ===")
nan_counts = df.isna().sum()
print(nan_counts)

if nan_counts.sum() > 0:
    print("⚠️  Il y a des NaN → vérifier / nettoyer.")
else:
    print("✓ Aucun NaN détecté.")
print()


# --- Statistiques de base ---
print("=== Statistiques descriptives ===")
print(df.describe())
print()


# --- Outliers potentiels ---
print("=== Vérification visuelle des outliers ===")
df[['glacier_mass_change_gt', 'glacier_area_km2']].boxplot(figsize=(10,5))
plt.title("Boxplots — vérifier les valeurs extrêmes anormales")
plt.show()
print("Regarde si les points extrêmes semblent impossibles (ex : valeurs énormes).")
print()


# --- Distribution des variables ---
print("=== Distribution des variables numériques ===")
df.hist(figsize=(12,10))
plt.show()
print("Regarde si une variable a une distribution étrange (pics extrêmes, valeurs isolées).")
print()


# --- Corrélation avec la variable cible ---
print("=== Corrélations avec glacier_mass_change_gt ===")
corrs = df.corr(numeric_only=True)['glacier_mass_change_gt'].sort_values(ascending=False)
print(corrs)
print("Si toutes les corrélations sont très faibles, le modèle apprendra peu.")
print()


# --- Doublons ---
print("=== Vérification des doublons ===")
dup_count = df.duplicated().sum()
print("Nombre de lignes dupliquées :", dup_count)

if dup_count > 0:
    print("⚠️  Il y a des doublons → probablement à supprimer.")
else:
    print("✓ Aucun doublon.")
print()


# --- Distribution de la cible ---
print("=== Distribution de la variable cible ===")
df["glacier_mass_change_gt"].hist(bins=50)
plt.title("Distribution de glacier_mass_change_gt")
plt.show()
print("Si distribution très déséquilibrée → attention aux performances du modèle.")
```

```python
import cdsapi

c = cdsapi.Client()

years = list(range(1975, 1977))
variables = ['2m_temperature','total_precipitation','surface_solar_radiation_downwards']

for year in years:
    c.retrieve(
        'reanalysis-era5-single-levels',
        {
            'product_type':'reanalysis',
            'variable':variables,
            'year':[str(year)],
            'month':[f"{m:02d}" for m in range(1,13)],
            'day':['01'],        # seulement le 1er du mois pour moyenne approximative
            'time':['00:00','06:00','12:00','18:00'],
            'area':[90, -180, -90, 180],  # ou filtrer plus précisément si souhaité
            'format':'netcdf',
        },
        f'era5_{year}.nc'
    )
```

```python

import xarray as xr

dfs = []

for year in years:
    ds = xr.open_dataset(f'era5_{year}.nc', engine='netcdf4')

    # Moyenne annuelle pour chaque lat/lon
    df_year = ds.mean(dim='time').to_dataframe().reset_index()

    # Conversion unités
    df_year['temperature'] = df_year['2m_temperature'] - 273.15  # K → °C
    df_year['precipitation'] = df_year['total_precipitation'] * 1000  # m → mm
    df_year['solar_radiation'] = df_year['surface_solar_radiation_downwards'] / 3.6e6  # J/m² → kWh/m²

    df_year = df_year[['year','lat','lon','temperature','precipitation','solar_radiation']]
    dfs.append(df_year)

df_climate = pd.concat(dfs, ignore_index=True)
print(df_climate.head())
print(df_climate.info())
```
